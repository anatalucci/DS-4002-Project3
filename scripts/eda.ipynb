# EDA

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/metadata")
image_dir = "data/images"

df["image_path"] = df["image_id"].apply(
    lambda x: os.path.join(image_dir, x + ".jpg")
)

df.head()
```

# Class Distribution

```python
df["dx"].value_counts().plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Types of Pigmented Lesions")
plt.ylabel("Count")
plt.savefig("../outputs/EDA/class_distribution.png", bbox_inches="tight")
plt.close()
```

# Age Distribution

```python
plt.hist(df["age"].dropna(), bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("../outputs/EDA/age_distribution.png", bbox_inches="tight")
plt.close()
```

# Sex Distribution

```python
df["sex"].value_counts().plot(kind="bar")
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.savefig("../outputs/EDA/sex_distribution.png", bbox_inches="tight")
plt.close()
```

# Localization Distribution

```python
df["localization"].value_counts().plot(kind="bar")
plt.title("Lesion Localization")
plt.xlabel("Localization")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.savefig("../outputs/EDA/localization_distribution.png", bbox_inches="tight")
plt.close()
```

# Heatmap

```python
pivot = pd.crosstab(df["localization"], df["dx"])

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Localization vs Diagnosis Heatmap")
plt.xlabel("Diagnosis (dx)")
plt.ylabel("Localization")

plt.tight_layout()
plt.savefig("../outputs/EDA/localization_vs_dx_heatmap.png", bbox_inches="tight")
plt.close()
```

```python
pivot = pd.crosstab(df["age"], df["dx"])

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Age vs Diagnosis Heatmap")
plt.xlabel("Diagnosis (dx)")
plt.ylabel("Age")

plt.tight_layout()
plt.savefig("../outputs/EDA/age_vs_dx_heatmap.png", bbox_inches="tight")
plt.show()
```

```python
pivot = pd.crosstab(df["sex"], df["dx"])

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Localization vs Diagnosis Heatmap")
plt.xlabel("Diagnosis (dx)")
plt.ylabel("Sex")

plt.tight_layout()
plt.savefig("../outputs/EDA/sex_vs_dx_heatmap.png", bbox_inches="tight")
plt.show()
```
