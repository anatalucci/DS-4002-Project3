# EDA

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
df = pd.read_csv("data/metadata")
image_dir = "data/images"

df["image_path"] = df["image_id"].apply(
    lambda x: os.path.join(image_dir, x + ".jpg")
)

df.head()
```

# Class Distribution

```python
df["dx"].value_counts().plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Types of Pigmented Lesions")
plt.ylabel("Count")
plt.savefig("EDA/class_distribution.png", bbox_inches="tight")
plt.show()
```

# Age Distribution

```python
plt.hist(df["age"].dropna(), bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("EDA/age_distribution.png", bbox_inches="tight")
plt.show()
```

# Sex Distribution

```python
df["sex"].value_counts().plot(kind="bar")
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.savefig("EDA/sex_distribution.png", bbox_inches="tight")
plt.show()
```

# Localization Distribution

```python
df["localization"].value_counts().plot(kind="bar")
plt.title("Lesion Localization")
plt.xlabel("Localization")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.savefig("EDA/localization_distribution.png", bbox_inches="tight")
plt.show()
```

# Sample Images

```python
sample_paths = df["image_path"].iloc[:5]

for path in sample_paths:
    img = Image.open(path)
    plt.imshow(img)
    plt.title(path.split("/")[-1])
    plt.axis("off")
for i, path in enumerate(sample_paths):
    img = Image.open(path)

    plt.imshow(img)
    plt.title(path.split("/")[-1])
    plt.axis("off")
    
    plt.savefig(f"EDA/sample_images/sample_{i}.png", bbox_inches="tight")
    plt.show()
```
